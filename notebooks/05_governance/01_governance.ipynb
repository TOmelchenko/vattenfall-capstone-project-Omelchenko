## Intended permissions

> Note: GRANT and SHOW GRANTS require metastore admin privileges not available in the trial workspace.
> In a production workspace, you would run the following:

```sql
-- catalog access
GRANT USE CATALOG ON CATALOG vattenfall_dev TO `analysts`;
GRANT USE CATALOG ON CATALOG vattenfall_dev TO `data_engineers`;

-- schema access
GRANT USE SCHEMA ON SCHEMA vattenfall_dev.analytics TO `analysts`;
GRANT USE SCHEMA ON SCHEMA vattenfall_dev.refined TO `data_engineers`;
GRANT USE SCHEMA ON SCHEMA vattenfall_dev.raw TO `data_engineers`;

-- table access
GRANT SELECT ON TABLE vattenfall_dev.analytics.gold_regional_operations TO `analysts`;
GRANT SELECT ON TABLE vattenfall_dev.analytics.gold_market_volatility TO `analysts`;
GRANT SELECT ON VIEW vattenfall_dev.analytics.vw_operational_dashboard TO `analysts`;

-- volume access
GRANT READ VOLUME ON VOLUME vattenfall_dev.raw.landing TO `data_engineers`;
GRANT WRITE VOLUME ON VOLUME vattenfall_dev.raw.landing TO `data_engineers`;
```
